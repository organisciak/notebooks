## Strip Nouns

This code will read each line of an input file and filter out anything that is not a noun. If `parse_mallet_input` is selected, it will present the name and label at the start.

It is useful for certain types of Topic Modelling (e.g. with MALLET)

```python
#@markdown First upload your input file to the *Files* tab on the left 
import spacy
nlp = spacy.load('en_core_web_sm')

input_doc_name = 'lyrics.txt' #@param {type:'string'}
output_doc_name = 'lyrics-cleaned.txt' #@param {type:'string'}
parse_mallet_input = True #@param {type:"boolean"}

with open(input_doc_name) as indoc, open(output_doc_name, mode='w') as outdoc:
    for line in indoc.readlines():
        if parse_mallet_input:
            try:
                name, label, line = line.split('\t', 2)
            except:
                name, label, line = line.split(' ', 2)
        doc = nlp(line, disable=['ner', 'lemmatizer', 'textcat'])
        nouns = [str(word) for word in doc if word.pos_ is 'NOUN']
        out_txt = " ".join(nouns) + '\n'
        if parse_mallet_input:
            out_txt = name + '\t' + label + '\t' + out_txt
        outdoc.write(out_txt)
```
